```python
import numpy as np
import matplotlib.pyplot as plt


import sys
sys.path.append('../class')
from NSGA2 import NSGA2
```

```python
# 设置初始参数
pop_size = 20 # 种群数量，需要为偶数才能两两组合进行交叉
max_gen = 5 # 最大迭代次数
min_x = -10 # x的最小值
max_x = 10 # x的最大值
```

```python
# 多目标的两个函数
def function1(x):
    return -x**2

def function2(x):
    return -(x-2)**2
```

```python
optimizer = NSGA2(func1=function1, func2=function2, pop_size=10, max_gen=1, prob_cross=0.9, prob_mut=0.1, min_x=min_x, max_x=max_x, random_seed=10)
optimizer.P
```

```python
optimizer.run(10)
optimizer.P
```

```python
plt.scatter(optimizer.values1_P, optimizer.values2_P)
for i in range(len(optimizer.P)):
    plt.text(optimizer.values1_P[i], optimizer.values2_P[i], str(i))
plt.xlabel('Function 1')
```
